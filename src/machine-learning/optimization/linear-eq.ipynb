```python
import numpy as np
```

```python
n = 10
m = 10
A = np.random.randn(n, m)
A += np.diag(np.sign(np.diag(A)) * 5)
b = np.random.randn(n)

print A
x0 = np.linalg.solve(A, b)
x0
```

```python
D = np.diag(A) 
B = A - np.diag(D)
DB = np.matmul(np.diag(1.0/D), B)
B
```

```python
x = np.random.rand(m)

for it in range(10):
    x = - np.matmul(DB, x) + 1.0/D * b
    print 'iter', it, 'error', np.linalg.norm(x - x0)
print x
```
