#### 因子分析

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
%config InlineBackend.figure_format = 'svg'

from sklearn.datasets import load_iris
```

```python
iris = load_iris()
data = iris.data
target = iris.target
```

```python
target
```

```python
from sklearn.decomposition import FactorAnalysis
fa = FactorAnalysis(3)
data_factors = fa.fit_transform(data)

for t, c in zip(range(3), list('rgb')):
    idx = target == t
    plt.plot(data_factors[idx,0], data_factors[idx,1], '%s.'% c)
```

```python
from sklearn.decomposition import PCA
pca = PCA(3)
data_pca = pca.fit_transform(data)

for t, c in zip(range(3), list('rgb')):
    idx = target == t
    plt.plot(data_pca[idx,0], data_pca[idx,1], '%s.' % c)
```
