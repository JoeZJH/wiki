```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.randn(128)
```

```python
plt.plot(x)
```

```python
x_fft = np.fft.fft(x)
```

```python
j = complex(0, 1)
def dft(xin, k):
    N = len(xin)
    if N == 1:
        return xin[0]
    
    xodd = xin[1::2]
    xeven = xin[::2]
    return np.exp(- j * 2 * np.pi / N * k) * dft(xodd, k) + dft(xeven, k)
```

```python
dft(x, 2)
```
